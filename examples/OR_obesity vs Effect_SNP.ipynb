```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../genocode/data/Obesity OR-Effect.csv")
```

```python
df
```

```python
plt.scatter(df["EPIC-Obesity Effect"].astype(np.float),df["OR Obesity"].astype(np.float))
```

```python
plt.scatter(df["EPIC-Obesity Effect"].astype(np.float),df["OR Obesity"].astype(np.float))
plt.errorbar(df["EPIC-Obesity Effect"].astype(np.float), df["OR Obesity"].astype(np.float), yerr=df["OR Obesity"].astype(np.float), label='both limits (default)')
z = np.polyfit(df["EPIC-Obesity Effect"].astype(np.float), df["OR Obesity"].astype(np.float), 1)
p = np.poly1d(z)
plt.plot(df["EPIC-Obesity Effect"].astype(np.float),p(df["EPIC-Obesity Effect"].astype(np.float)),"r--")
print(p)
```

```python
plt.scatter(df["OR Obesity"].astype(np.float),df["EPIC-Obesity Effect"].astype(np.float))
plt.errorbar(df["OR Obesity"].astype(np.float),df["EPIC-Obesity Effect"].astype(np.float), yerr=df["EPIC-Obesity Serr"].astype(np.float), linestyle="None", uplims=True, lolims=True)
z = np.polyfit(df["OR Obesity"].astype(np.float),df["EPIC-Obesity Effect"].astype(np.float), 1)
p = np.poly1d(z)
plt.plot(df["OR Obesity"].astype(np.float),p(df["OR Obesity"].astype(np.float)),"r--")
plt.xlabel("OR Obesity", size=12)
plt.ylabel("EPIC-Obesity Effect", size=12)
plt.rcParams["figure.figsize"] = (10,8)
plt.suptitle('OR of obesity vs Effect of SNP', size=18)
plt.title("y=%.6fx+%.6f"%(z[0],z[1]), size=15) 
print(p)
```

```python
plt.scatter(df["EPIC-Obesity Effect"],df["OR Obesity"])
plt.errorbar(df["EPIC-Obesity Effect"],df["OR Obesity"], linestyle="None", uplims=True, lolims=True)
z = np.polyfit(df["EPIC-Obesity Effect"],df["OR Obesity"], 2)
p = np.poly1d(z)
plt.plot(df["EPIC-Obesity Effect"],p(df["EPIC-Obesity Effect"]),"r--")
plt.xlabel("EPIC-Obesity study, Allele Effect", size=12)
plt.ylabel("OR Obesity", size=12)
plt.rcParams["figure.figsize"] = (10,8)
plt.suptitle('')
plt.title("y=%.6fx+%.6f"%(z[0],z[1]), size=15) 
print(p)
```
